```python
import pandas as pd
```

```python
nobel = pd.read_csv("nobel.csv")
```

```python
nobel.head(5)
```

```python
winner = nobel[["Laureate ID", "Full Name", "Birth Date", "Sex"]].copy()
new_columns = {"Laureate ID": "Laureate_ID", "Full Name": "Full_Name", "Birth Date": "Birth_Date", "Sex":"Sex"}
winner.rename(columns=new_columns, inplace=True)
```

```python
winner
```

```python
gpd = pd.read_csv("gpd.csv")
c_to_drop = []
for r in range(1960,1991):
    c_to_drop.append(str(r))
for c in gpd.columns:
    if c in c_to_drop:
        gpd.drop(c, axis = 1, inplace=True)
gpd.drop(["Country Code", "Indicator Name", "Indicator Code", "Unnamed: 67"], axis = 1, inplace=True)
gpd.rename(columns={'Country Name': 'Country_Name', "Indicator Name": "Indicator_Name"}, inplace=True)
gpd
```

```python
gpd.to_csv("gpd_cleaned.csv")
```

```python
nationality = nobel[["Laureate ID", "Birth City", "Birth Country"]].copy()
new_columns = {"Laureate ID": "Laureate_ID", "Birth City": "Birth_City", "Birth Country": "Birth_Country"}
nationality.rename(columns=new_columns, inplace=True)
nationality
```

```python
nationality.to_csv("nationality.csv")
```

```python
winner.to_csv("winner.csv")
```

```python
prize = nobel[["Laureate ID", "Year", "Category", "Laureate Type"]].copy()
new_columns = {"Laureate ID": "Laureate_ID", "Year": "Year", "Category": "Category", "Laureate Type": "Laureate_Type"}
prize.rename(columns=new_columns, inplace=True)
```

```python
prize
```

```python
prize.to_csv("prize.csv")
```

```python
organization = nobel[["Laureate ID", "Organization Name", "Organization Country", "Organization City"]].copy()
new_columns = {"Laureate ID": "Laureate_ID", "Organization Name": "Organization_Name", "Organization Country": "Organization_Country", "Organization City": "Organization_City"}
organization.rename(columns=new_columns, inplace=True)
organization
```

```python
organization.to_csv("organization.csv")
```

```python
winner_with_org = pd.merge(winner, organization, on='Laureate_ID', how='left')

winner_with_org['Organization'] = winner_with_org['Organization_Name'].notna().astype(int)

winner_with_org
```

```python
winner_with_org.to_csv("winner_org.csv")
```
